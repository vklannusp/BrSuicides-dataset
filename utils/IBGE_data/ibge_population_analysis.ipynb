```python
import os
import pandas as pd
from IPython.display import display

# Diretório onde o CSV está localizado
dir_path = './final_all_years_result_csv'
# Nome do arquivo CSV na pasta
csv_file_name = [f for f in os.listdir(dir_path) if f.endswith('.csv')][0]
# Caminho completo do arquivo
csv_file_path = os.path.join(dir_path, csv_file_name)
# Leitura do CSV
dataframe = pd.read_csv(csv_file_path, index_col=0)

dataframe[dataframe.index == 'Brasil']
```

```python
# No brackets inside .loc[id], locating cols that correspond to "Brasil"
br_data = dataframe.loc["Brasil"]

growth_rates = br_data.pct_change() * 100
growth_rate_col = "Crescimento Anual (%)"
growth_rates_df = pd.DataFrame([growth_rates], index=[growth_rate_col])

display(dataframe.loc[["Brasil"]])

# Brackets inside .loc[[id]]), to get the row "Brasil"
result_df = pd.concat([dataframe.loc[["Brasil"]], growth_rates_df])
result_df.loc[growth_rate_col] = result_df.loc[growth_rate_col].map("{:.2f}".format)


# Exibindo o resultado
display(result_df.T)
```
